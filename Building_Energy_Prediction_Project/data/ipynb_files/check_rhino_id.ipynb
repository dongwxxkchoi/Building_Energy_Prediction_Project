```python
import rhino3dm
```

```python
rhino_file = '../../3.rhino/samsung_building_modeling.3dm'

# Rhino 파일 열기
model = rhino3dm.File3dm.Read(rhino_file)

# 다각형(Polygon)에 대한 ID 조회
for geometry in model.Objects:
    print(geometry)
    if isinstance(geometry.Geometry, rhino3dm.Polyline):
        polygon_id = geometry.Attributes.ObjectId
        print(f"Polygon ID: {polygon_id}")
```
