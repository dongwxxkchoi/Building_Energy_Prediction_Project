```python
cursor.execute("SELECT * FROM nonplottable_setting")
rows = cursor.fetchall()

ret_list = []
res_list = []
off_list = []


# template 별 id 수록
for row in rows:
    if row[2] == 'TemplateName':
        if row[3] == 'Boston_Ret_0':
            ret_list.append(row[1])
        if row[3] == 'Boston_Res_0_Masonry':
            res_list.append(row[1])
        if row[3] == 'Boston_Off_0':
            off_list.append(row[1])
```

```python
print(len(ret_list))
print(len(res_list))
print(len(off_list))
```
