```python
import sqlite3

# 데이터베이스 연결
db_path = "../../3.rhino/samsung_test_open/umi.sqlite3"
conn = sqlite3.connect(db_path)

cursor = conn.cursor()

# 모든 테이블명 조회
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cursor.fetchall()

# 조회된 테이블명 출력
for table in tables:
    print(table[0])

# 데이터베이스 연결 종료
conn.close()
```
